# MFI vs RLU for different clones

```python
import os
import warnings

import math
import numpy as np 

from IPython.display import display, HTML
import matplotlib.pyplot as plt

from neutcurve.colorschemes import CBMARKERS, CBPALETTE

import pandas as pd
from plotnine import *

import yaml
```

```python
MFI_RLU_df = pd.read_csv('MFI_vs_RLU.csv')
MFI_RLU_df
```

```python
p = (ggplot(MFI_RLU_df, 
            aes(x='relative_MFI', y='relative_RLU', color='cells')
           ) +
     geom_point(size=2.5)+
     theme_classic() +
     theme(axis_text_x=element_text(angle=90, vjust=1, hjust=1),
           axis_title_x=element_blank(),
           figure_size=(4,4)
          ) +
          scale_y_log10(limits=[0.01,1]) +
        scale_x_log10(limits=[0.01,1]) +
     labs(x = 'relative Ace2 expression',
          y ='relative infection',
          title ='ACE2 expression versus cell infectivety') +
     scale_color_manual(values=("#009E73","#56B4E9","#E69F00","#999999"))
    )

p = p + xlab('relative Ace2 expression')
_ = p.draw()

p.save('MFI_vs_RLU.pdf')
```

```python
p = (ggplot(MFI_RLU_df, 
            aes(x='cells', y='RLU/ul', color='cells')
           ) +
     geom_point(size=2.5)+
     theme_classic() +
     theme(axis_text_x=element_text(angle=90, vjust=1, hjust=1),
           axis_title_x=element_blank(),
           figure_size=(4,4)
          ) +
          scale_y_log10(limits=[1,1.1e6]) +
     labs(y ='relative infection',
          title ='Cell clone versus infectivety') +
     scale_color_manual(values=("#009E73","#56B4E9","#E69F00","#999999"))
    )

p = p + xlab('relative Ace2 expression')
_ = p.draw()

p.save('cell_vs_RLU.pdf')
```

```python
p = (ggplot(MFI_RLU_df, 
            aes(x='cells', y='MFI', color='cells')
           ) +
     geom_point(size=2.5)+
     theme_classic() +
     theme(axis_text_x=element_text(angle=90, vjust=1, hjust=1),
           axis_title_x=element_blank(),
           figure_size=(4,4)
          ) +
          scale_y_log10(limits=[1,1.1e6]) +
     labs(y ='mean fluorescence intensity',
          title ='Cell clone versus ACE2 expression') +
     scale_color_manual(values=("#009E73","#56B4E9","#E69F00","#999999"))
    )

_ = p.draw()

p.save('cell_vs_MFI.pdf')

```
